# **3. Remove URLs**

```python
def remove_urls(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)
```

# **4. Remove Punctuations **

```python
import string, time
punctuations= string.punctuation
punctuations
```

```python
def remove_punc(text):
    for char in punctuations:
        text= text.replace(char, '')
    return text
```

```python
text = 'string# with. Punctuations? \\'
remove_punc(text)
```

```python
# Punctuation remove Faster
def remove_punc1(text):
    return text.translate(str.maketrans('', '', punctuations))
```

```python
text = 'string# with. Punctuations? \\'
remove_punc1(text)
```

```python
data['review']=data['review'].apply(remove_punc1)
```

# **8. Emoji Handling**
1. remove them
2.replace with its meaning (ðŸ˜‚ = laughing)

```python
import re
def remove_emoji(text):
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emotions
        u"\U0001F300-\U0001F5FF"  # symbols and pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (ios)
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

# **10. Stemming**
Prcess of reducing words to their root or base form by removing suffixes or prefixes, stemming doesn't consider grammatical correctness.
e.g.
walk - walk
walks - walk
walking - walk
walked - walk

grammatical mistakes occur commonly.
e.g.
study - study
studies - studi
studied - studi
studying - studyi* 

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample = 'walk walks walking walked study studies studying studied'
stem_words(sample)
```

```python
print(data['review'][0])
print("")
print(stem_words(data['review'][0]))
```

# **11. Lemmatization**
Lemmatization is the process of reducing words to their base or dictionary form (lemma) by considering grammar, part-of-speech, and meaning.
It is slower than stemming.
Lemmatization uses lexical resources such as WordNet (WordNetLemmatizer) to obtain the correct root form of a word.

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
wordnet_lemmatizer= WordNetLemmatizer()
```

```python
sent = data['review'][1]
print(sent)
```

```python
tokenize= word_tokenize(sent)
print(tokenize)
```

```python
lemmatized_sent1 = [wordnet_lemmatizer.lemmatize(word) for word in tokenize]
print(lemmatized_sent1)
```

```python
lemmatized_sent2 = [wordnet_lemmatizer.lemmatize(word, pos='v') for word in tokenize]
print(lemmatized_sent2)
```
