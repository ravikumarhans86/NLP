# **2. Remove HTML tags**

```python
import re
def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub('r', text)
```

# **3. Remove URLs**

```python
def remove_urls(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)
```

```python
def chat_conversion(text):
    new_text= []
    for w in text.split():
        if w.upper() in chat_word:
            new_text.append(chat_word[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

# **7. Stop Word Removing**

```python
from nltk.corpus import stopwords
stop_words=stopwords.words('english')
```

```python
def remove_stopword(text):
    ans = " ".join(w for w in text.split() if w not in stop_words)
    return ans
```

# **8. Removing or replace Emoji**

```python
import re
def remove_emoji(text):
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emotions
        u"\U0001F300-\U0001F5FF"  # symbols and pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (ios)
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

# **12.  No of Words entore Corpus**

```python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')

total_words = 0

for text in df['review'].dropna():
    tokens = word_tokenize(text)
    total_words += len(tokens)

print("Total words in corpus:", total_words)
```

```python
# Total no of words
total_words = 0

for text in df['review'].dropna():
    tokens = word_tokenize(text.lower())
    tokens = [w for w in tokens if w.isalpha()]
    total_words += len(tokens)

print("Cleaned total words:", total_words)
```

```python
#Unique Words for Vocabulary
vocab = set()

for text in df['review'].dropna():
    tokens = word_tokenize(text.lower())
    tokens = [w for w in tokens if w.isalpha()]
    vocab.update(tokens)

print("Vocabulary size:", len(vocab))
```
